```python
import numpy as np
import random
import torch
import torch.nn as nn
import torch.optim as optim
import gym
from collections import deque
import matplotlib.pyplot as plt
import pandas as pd

# Define the RL Environment for Construction Logistics
class ConstructionLogisticsEnv(gym.Env):
    def __init__(self):
        super(ConstructionLogisticsEnv, self).__init__()

        self.truck_capacity = 10                   # Each truck can deliver 10 tons
        self.valid_order_quantities = [0, 10, 20, 30, 40]  # Only full truckloads allowed
        self.max_inventory = 60                    # Site can store 60 tons max
        self.max_order = max(self.valid_order_quantities)  # Maximum possible per day
        
        self.storage_cost = 2                     # £2/ton/day (premium urban cost)
        self.shortage_cost = 100                  # £100/ton for delay-related downtime
        self.transport_cost = 100                 # £100/truck (fixed per truck, not per ton)
        
        self.time_horizon = 11
        self.current_time = 0
        self.inventory_level = 0
        self.lead_time = []
        self.pending_orders = deque()
        self.backlog = 0
        self.debug = False
        self.recent_truck_usage = [0, 0]  # usage on t-1 and t-2
        
        # Observation space adjusted for batching logic and max capacity
        # State = [inventory, day, backlog, expected_demand_today, lead_time, expected_demand_tomorrow, trucks used on day t-1 and t-2]
        self.observation_space = gym.spaces.Box(
            low=np.array([0, 0, 0, 0, 1, 0, 0, 0], dtype=np.float32),
            high=np.array([
                self.max_inventory,       # inventory
                self.time_horizon,        # current time
                2 * self.max_inventory,   # backlog
                20,                       # expected demand
                2,                        # lead time
                20,                       # next day demand
                4,                        # trucks used on day t-1
                4                        # trucks used on day t-2
            ], dtype=np.float32),
            dtype=np.float32
        )

        # Updated action space to reflect only batch sizes
        self.action_space = gym.spaces.Discrete(len(self.valid_order_quantities))
      

    def step(self, action):
        # Receive arriving orders
        while self.pending_orders and self.pending_orders[0][1] == self.current_time:
            delivery_qty = self.pending_orders.popleft()[0]  # ← YOU NEED THIS LINE
            available_space = self.max_inventory - self.inventory_level
            actual_delivery = min(delivery_qty, available_space)
            self.inventory_level += actual_delivery#
            if self.debug:
                print("Day: " ,self.current_time," Today's starting inventory: ", self.inventory_level)
            
        # Generate random demand influences
        self.weather_factor = np.random.choice([0.6, 0.85, 1.0], size=self.time_horizon, p=[0.02, 0.1, 0.88])
        self.crane_breakdown = np.random.choice([0.5, 1.0], size=self.time_horizon, p=[0.02, 0.98])
        self.worker_productivity = np.random.choice([0.8, 1.0], size=self.time_horizon, p=[0.05, 0.95])
        self.trnasport_latness = np.random.choice([0.8, 0.9, 1.0], size = self.time_horizon,p= [0.1,0.8, 0.1])
        
        # Generate stochastic demand
        #mean_demand, std_dev = 18, 4        
        raw_demand = self.expected_demand[self.current_time]       
        adjusted_demand = raw_demand * self.weather_factor[self.current_time] \
                                        * self.crane_breakdown[self.current_time] \
                                        * self.worker_productivity[self.current_time]\
                                        * self.trnasport_latness[self.current_time]
        if self.current_time == 0:
            demand = 0
        else:
            demand = int(adjusted_demand) + self.backlog
            #demand = 5 + self.backlog
        if self.debug:    
            print("Day: " ,self.current_time," Today's Demand: ", demand)
        
        # Fulfill demand       
        shortage = max(0, demand - self.inventory_level)
        if self.debug:
            print("Day: " ,self.current_time," Today's Unmet: ", shortage)
        fulfilled = demand - shortage
        if self.debug:
            print("Day: " ,self.current_time," Today's fulfilled: ", fulfilled)
        self.inventory_level -= fulfilled
        if self.debug:
            print("Day: " ,self.current_time," Today's Final inventory level: ", self.inventory_level)
        self.backlog = shortage  # carry unmet demand to tomorrow
        if self.debug:
            print("Day: " ,self.current_time," backlog: ", self.backlog)
        
        # Place order if valid
        order_quantity = self.valid_order_quantities[action]
        num_trucks_today = order_quantity // self.truck_capacity
        self.recent_truck_usage.append(num_trucks_today)
        self.recent_truck_usage = self.recent_truck_usage[-2:]
        if order_quantity > 0:
            delivery_time = self.current_time + self.lead_time[self.current_time]
            if delivery_time < self.time_horizon:
                self.pending_orders.append((order_quantity, delivery_time))
                if self.debug:
                    print(f"Day: {self.current_time}, Order quantity: {order_quantity}, Delivery at: Day {delivery_time}")
             #else:
            #    action = 0
    
        holding_cost = self.storage_cost * self.inventory_level
        if self.debug:
            print("Day:",self.current_time," holding cost: ", holding_cost)
        shortage_penalty = self.shortage_cost * shortage
        if self.debug:
            print("Day:",self.current_time," shortage_penalty: ", shortage_penalty)
        transport_cost = num_trucks_today * self.transport_cost
        if self.debug:
            print("Day:",self.current_time," transport_cost: ", transport_cost)
        
        # --- Base reward (negative cost) ---
        total_cost =  transport_cost + holding_cost + shortage_penalty 
        reward = -total_cost  # Default baseline reward
        if self.debug:
            print("Day:",self.current_time," Total cost today ", reward)
            print("====================================================")
            
         # --- Improved Reward Shaping (Just-in-Time Incentive) ---

        # --- JIT-Inspired Reward Shaping ---
        
        # 1. Bonus for exact match (ideal inventory = tomorrow's demand)
        if 0 < self.current_time < self.time_horizon - 1:
            ideal_inventory = self.expected_demand[self.current_time + 1]  # anticipate tomorrow
            diff = abs(self.inventory_level - ideal_inventory)
            reward += max(0, 20 - 2 * diff)  # bonus decreases as deviation grows
        
        # 2. Penalty for excessive stock (beyond what's needed for next 2 days)
        if self.current_time < self.time_horizon - 2:
            next_two_days_demand = self.expected_demand[self.current_time + 1] + self.expected_demand[self.current_time + 2]
            if self.inventory_level > next_two_days_demand + 5:
                penalty = (self.inventory_level - next_two_days_demand) * 1.5
                reward -= penalty
                if self.debug:
                    print(f"Overstock penalty: -{penalty:.1f}")
        
        # 3. End-of-horizon inventory penalty (already included, keep this)
        if self.current_time == self.time_horizon - 1:
            excess_penalty = self.inventory_level * 5
            reward -= excess_penalty
            if self.debug:
                print(f"End-of-episode penalty for leftover inventory: -{excess_penalty}")
        
        # 4. JIT Bonus for low inventory AND no shortage
        if shortage == 0 and self.inventory_level <= 2:
            reward += 30
        
        # 5. Deterrent for ordering when inventory is already near max
        if action > 0 and self.inventory_level > self.max_inventory * 0.8:
            reward -= 10
            
        next_day_demand = self.expected_demand[self.current_time + 1] if self.current_time + 1 < self.time_horizon else 0
        
        self.current_time += 1
        done = self.current_time >= self.time_horizon
    
        if done:     
            return np.zeros(self.observation_space.shape, dtype=np.float32), reward, done, {"cost": total_cost}
    
        state = np.array([
            self.inventory_level,                      # Current inventory
            self.current_time,                         # Current day
            self.backlog,                              # Unfulfilled demand
            self.expected_demand[self.current_time],   # Based on planned site activities
            self.lead_time[self.current_time],          # Current lead time
            next_day_demand,
            self.recent_truck_usage[0],
            self.recent_truck_usage[1]
        ], dtype=np.float32)
    
        return state, reward, done, {"cost": total_cost}


    def reset(self):
        self.current_time = 0
        self.inventory_level = 0
        self.backlog = 0        
        self.lead_time = np.random.choice([1, 2], size=self.time_horizon, p=[0.9, 0.1])
        self.pending_orders = deque()
        #self.expected_demand = np.full(self.time_horizon, 10)
        self.expected_demand = np.array([12, 14, 10, 8, 18, 20, 10, 10, 14, 10, 12])
        next_day_demand = self.expected_demand[1] if self.time_horizon > 1 else 0
        self.recent_truck_usage = [0, 0]  # Reset truck usage history
    
        state = np.array([
            self.inventory_level,
            self.current_time,
            self.backlog,
            self.expected_demand[self.current_time],
            self.lead_time[self.current_time],
            next_day_demand,
            self.recent_truck_usage[0],
            self.recent_truck_usage[1]
        ], dtype=np.float32)
        
        return state

class DQN(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(DQN, self).__init__()
        self.fc1 = nn.Linear(input_dim, 128)
        self.fc2 = nn.Linear(128, 128)
        self.fc3 = nn.Linear(128, output_dim)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        return self.fc3(x)

env = ConstructionLogisticsEnv()
dqn = DQN(input_dim=8, output_dim=env.action_space.n)
target_dqn = DQN(input_dim=8, output_dim=env.action_space.n)
target_dqn.load_state_dict(dqn.state_dict())

optimizer = optim.Adam(dqn.parameters(), lr=0.0001)
loss_fn = nn.MSELoss()

epsilon = 1.0
epsilon_decay = 0.997 #0.991 #0.997
epsilon_min = 0.05 #0.05
num_episodes = 8000
batch_size = 64
memory = deque(maxlen=50000)
soft_tau = 0.05

ema_alpha = 0.05

def ema(values, alpha):
    ema_values = []
    for i, v in enumerate(values):
        if i == 0:
            ema_values.append(v)
        else:
            ema_values.append(alpha * v + (1 - alpha) * ema_values[-1])
    return ema_values

rewards_per_episode = []
losses_per_episode = []
q_variances = []

shaped_rewards_per_episode = []
true_costs_per_episode = []

def choose_action(state):
    # Get the current inventory level
    inventory_level = state[0]
    
    # Only allow actions that don't exceed max inventory
    possible_actions = [    i for i, qty in enumerate(env.valid_order_quantities) if inventory_level + qty <= env.max_inventory]

    # If the inventory is full, only allow action 0 (no order)
    if inventory_level >= env.max_inventory:
        possible_actions = [0]

    #If no valid actions, default to ordering 0
    #if not possible_actions:
        #print(f"No valid actions available for inventory level {inventory_level}. Choosing action 0.")
        #return 0
            
    # If epsilon-greedy, choose randomly from the valid actions
    if random.random() < epsilon:
        return random.choice(possible_actions)
    
    # Otherwise, choose the best action from the valid actions
    with torch.no_grad():
        q_values = dqn(torch.tensor(state, dtype=torch.float32).unsqueeze(0)).squeeze(0)
        valid_q_values = [q_values[action] for action in possible_actions]
        best_action = possible_actions[torch.argmax(torch.tensor(valid_q_values)).item()]
        return best_action

#visitation = np.zeros((env.max_inventory + 1, env.time_horizon + 1), dtype=int)

for episode in range(num_episodes):

    if episode == num_episodes-1:
        env.debug = True  # Turn on detailed logging
    else:
        env.debug = False  # Turn off for all other episodes
    
    state = env.reset()
    total_reward = 0
    done = False
    total_cost = 0  # New: for real-world cost

    while not done:
        #inv_level = int(state[0])
        #days = int(state[1])
        #visitation[inv_level, days] += 1
        
        action = choose_action(state)
        next_state, reward, done, _ = env.step(action)
        real_cost = -reward if "cost" not in _ else _["cost"]
        
        memory.append((state, action, reward, next_state, done))
        state = next_state
        total_reward += reward
        total_cost += real_cost


        
    rewards_per_episode.append(total_reward)
        
    shaped_rewards_per_episode.append(total_reward)
    true_costs_per_episode.append(total_cost)

    if len(memory) > batch_size:
        batch = random.sample(memory, batch_size)
        states, actions, rewards, next_states, dones = zip(*batch)

        states = torch.tensor(states, dtype=torch.float32)
        actions = torch.tensor(actions, dtype=torch.int64)
        rewards = torch.tensor(rewards, dtype=torch.float32)
        next_states = torch.tensor(next_states, dtype=torch.float32)
        dones = torch.tensor(dones, dtype=torch.bool)

        q_values = dqn(states).gather(1, actions.unsqueeze(1)).squeeze(1)
        next_q_values = target_dqn(next_states).max(1)[0].detach()
        target_q_values = rewards + (0.99 * next_q_values * (~dones))
        loss = loss_fn(q_values, target_q_values)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Soft update
        for target_param, param in zip(target_dqn.parameters(), dqn.parameters()):
            target_param.data.copy_(soft_tau * param.data + (1.0 - soft_tau) * target_param.data)

        losses_per_episode.append(loss.item())
        q_variances.append(dqn(states).var().item())
   
    epsilon = max(epsilon * epsilon_decay, epsilon_min)

    if episode % 100 == 0:
        
        print(f"Episode {episode}, Total Reward: {total_reward}")
        
# Apply EMA smoothing
rewards_ema = ema(rewards_per_episode, ema_alpha)
loss_ema = ema(losses_per_episode, ema_alpha)
q_var_ema = ema(q_variances, ema_alpha)

# Plot Performance
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(rewards_per_episode, alpha=0.3, label='Raw')
plt.plot(rewards_ema, label='EMA Smoothed')
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.title('Training Reward')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(losses_per_episode, alpha=0.3, label='Raw')
plt.plot(loss_ema, label='EMA Smoothed')
plt.xlabel('Episode')
plt.ylabel('Bellman Error (Loss)')
plt.title('Bellman Error Over Time')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(q_variances)
plt.plot(q_var_ema, label='Q-Variance EMA')
plt.xlabel('Episode')
plt.ylabel('Q-Value Variance')
plt.title('Q-Value Variance Over Time')

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(shaped_rewards_per_episode, label="Shaped Reward (used for training)")
plt.plot([-c for c in true_costs_per_episode], label="True Total Cost (negated)", linestyle="--")
plt.xlabel("Episode")
plt.ylabel("Reward / -Cost")
plt.legend()
plt.title("Training Progress: Shaped Reward vs True Cost")
plt.grid(True)
plt.show()

torch.save(dqn.state_dict(), "trained_dqn_model.pth")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def extract_policy_table(env, dqn, num_days, max_inventory):
    policy_table = np.zeros((max_inventory + 1, num_days))

    for day in range(num_days):
        for inv_level in range(max_inventory + 1):
            state = np.array([
                inv_level,      # inventory_level
                day,            # current day
                0,              # backlog (assume 0)
                10,             # expected_demand_today (fixed)
                1,              # lead_time (fixed)
                10,             # expected_demand_tomorrow (same as today for simplicity)
                1,              # recent_truck_usage_t-1
                1               # recent_truck_usage_t-2
            ], dtype=np.float32)

            action = choose_action(state)
            order_qty = env.valid_order_quantities[action]
            policy_table[inv_level, day] = order_qty

    return pd.DataFrame(policy_table,
                        index=[f"Inv. {i}" for i in range(max_inventory + 1)],
                        columns=[f"Day {i}" for i in range(num_days)])
    
# Generate and save the policy heatmap
policy_df = extract_policy_table(env, dqn, num_days=11, max_inventory=60)
print(policy_df)

# Save to Excel
policy_df.to_excel("policy_table.xlsx", index=True)
print("Policy table has been saved to 'policy_table.xlsx'")

# Visualise heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(policy_df, annot=True, cmap="Blues", fmt=".0f",
            cbar_kws={'label': 'Order Quantity (tons)'},
            xticklabels=policy_df.columns, yticklabels=policy_df.index,
            annot_kws={"size": 8})

plt.title("Policy Heatmap: Optimal Daily Order Quantity (in tons)", fontsize=14)
plt.xlabel("Day", fontsize=12)
plt.ylabel("Inventory Level (tons)", fontsize=12)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
env.debug = False 
def evaluate_fixed_order_policy(env, order_quantity=20, episodes=100):
    total_costs = []

    for _ in range(episodes):
        state = env.reset()
        done = False
        total_cost = 0

        while not done:
            action = env.valid_order_quantities.index(order_quantity)
            _, _, done, info = env.step(action)
            total_cost += info["cost"]

        total_costs.append(total_cost)
    
    return np.mean(total_costs), np.std(total_costs)

def evaluate_threshold_policy(env, threshold=10, order_quantity=20, episodes=100):
    total_costs = []

    for _ in range(episodes):
        state = env.reset()
        done = False
        total_cost = 0

        while not done:
            inventory = state[0]
            if inventory < threshold:
                action = env.valid_order_quantities.index(order_quantity)
            else:
                action = 0  # no order
            state, _, done, info = env.step(action)
            total_cost += info["cost"]

        total_costs.append(total_cost)
    
    return np.mean(total_costs), np.std(total_costs)

def evaluate_myopic_policy(env, episodes=100):
    total_costs = []

    for _ in range(episodes):
        state = env.reset()
        done = False
        total_cost = 0

        while not done:
            inventory = state[0]
            demand_today = state[3]
            lead_time = int(state[4])

            # Estimate cost for each possible action (greedy)
            best_action = 0
            best_cost = float('inf')
            for i, order_qty in enumerate(env.valid_order_quantities):
                projected_inventory = inventory + order_qty
                shortage = max(0, demand_today - projected_inventory)
                holding = max(0, projected_inventory - demand_today)
                trucks = order_qty // env.truck_capacity
                cost = (
                    shortage * env.shortage_cost +
                    holding * env.storage_cost +
                    trucks * env.transport_cost
                )
                if cost < best_cost:
                    best_cost = cost
                    best_action = i

            state, _, done, info = env.step(best_action)
            total_cost += info["cost"]

        total_costs.append(total_cost)
    
    return np.mean(total_costs), np.std(total_costs)

#mean_cost_dqn = np.mean(true_costs_per_episode[-100:])
mean_cost_dqn = np.mean(true_costs_per_episode[-100:])
std_cost_dqn = np.std(true_costs_per_episode[-100:])
mean_cost_fixed, std_fixed = evaluate_fixed_order_policy(env)
mean_cost_thresh, std_thresh = evaluate_threshold_policy(env)
mean_cost_myopic, std_myopic = evaluate_myopic_policy(env)

print(f"DQN:    Mean Cost = {mean_cost_dqn:.2f}")
print(f"Fixed:  Mean = {mean_cost_fixed:.2f}, Std = {std_fixed:.2f}")
print(f"Thresh: Mean = {mean_cost_thresh:.2f}, Std = {std_thresh:.2f}")
print(f"Myopic: Mean = {mean_cost_myopic:.2f}, Std = {std_myopic:.2f}")


results_df = pd.DataFrame({
    "Policy": ["DQN", "Fixed Order", "Threshold (15)", "Myopic"],
    "Mean Cost": [mean_cost_dqn, mean_cost_fixed, mean_cost_thresh, mean_cost_myopic],
    #"Std. Dev": [0, std_fixed, std_thresh, std_myopic]  # DQN std = 0 as you used mean over last 100
    "Std. Dev": [std_cost_dqn, std_fixed, std_thresh, std_myopic]
})

print(results_df)

policies = ["DQN", "Fixed", "Threshold (10)", "Myopic"]
means = [mean_cost_dqn, mean_cost_fixed, mean_cost_thresh, mean_cost_myopic]
stds = [std_cost_dqn, std_fixed, std_thresh, std_myopic]

plt.figure(figsize=(10, 6))
bars = plt.bar(policies, means, yerr=stds, capsize=10)
plt.ylabel("Total Cost")
plt.title("Baseline Policy Comparison vs DQN")
plt.grid(axis='y')

# Highlight DQN bar
bars[0].set_color('green')

plt.tight_layout()
plt.show()
```

```python
env.debug = False 
def evaluate_shortage_incidents(env, policy_func, label, episodes=100):
    shortage_counts = []

    for _ in range(episodes):
        state = env.reset()
        done = False
        daily_shortage_count = 0

        while not done:
            action = policy_func(state, env)
            state, _, done, _ = env.step(action)
            if state[2] > 0:  # backlog > 0 implies shortage
                daily_shortage_count += 1

        shortage_counts.append(daily_shortage_count)

    return label, np.mean(shortage_counts), np.std(shortage_counts)

# Policy functions
def dqn_policy(state, env):
    inventory_level = state[0]
    possible_actions = [i for i, qty in enumerate(env.valid_order_quantities)
                        if inventory_level + qty <= env.max_inventory]
    if inventory_level >= env.max_inventory:
        possible_actions = [0]
    with torch.no_grad():
        q_values = dqn(torch.tensor(state, dtype=torch.float32).unsqueeze(0)).squeeze(0)
        valid_q_values = [q_values[action] for action in possible_actions]
        best_action = possible_actions[torch.argmax(torch.tensor(valid_q_values)).item()]
        return best_action

def fixed_policy(state, env):
    return env.valid_order_quantities.index(20)

def threshold_policy(state, env, threshold=10):
    return env.valid_order_quantities.index(20) if state[0] < threshold else 0

def myopic_policy(state, env):
    inventory = state[0]
    demand_today = state[3]
    best_action = 0
    best_cost = float('inf')
    for i, order_qty in enumerate(env.valid_order_quantities):
        projected_inventory = inventory + order_qty
        shortage = max(0, demand_today - projected_inventory)
        holding = max(0, projected_inventory - demand_today)
        trucks = order_qty // env.truck_capacity
        cost = (
            shortage * env.shortage_cost +
            holding * env.storage_cost +
            trucks * env.transport_cost
        )
        if cost < best_cost:
            best_cost = cost
            best_action = i
    return best_action

# Run evaluation
results = []
results.append(evaluate_shortage_incidents(env, dqn_policy, "DQN"))
results.append(evaluate_shortage_incidents(env, fixed_policy, "Fixed"))
results.append(evaluate_shortage_incidents(env, lambda s, e: threshold_policy(s, e, threshold=10), "Threshold (10)"))
results.append(evaluate_shortage_incidents(env, myopic_policy, "Myopic"))

# Plotting
labels, means, stds = zip(*results)

plt.figure(figsize=(10, 6))
bars = plt.bar(labels, means, yerr=stds, capsize=10)
plt.ylabel("Avg. Days with Shortage per Episode")
plt.title("Shortage Incidents: DQN vs Baseline Policies")
bars[0].set_color('green')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
env.debug = False 
def evaluate_shortage_trend(env, policy_func, label, episodes=100):
    shortage_trend = np.zeros(env.time_horizon)

    for _ in range(episodes):
        state = env.reset()
        done = False
        daily_shortages = np.zeros(env.time_horizon)
        t = 0

        while not done:
            action = policy_func(state, env)
            state, _, done, _ = env.step(action)
            if state[2] > 0:  # backlog > 0 implies shortage
                daily_shortages[t] = 1
            t += 1

        shortage_trend += daily_shortages

    average_trend = shortage_trend / episodes
    return label, average_trend

# Collect data
shortage_trends = []
shortage_trends.append(evaluate_shortage_trend(env, dqn_policy, "DQN"))
shortage_trends.append(evaluate_shortage_trend(env, fixed_policy, "Fixed"))
shortage_trends.append(evaluate_shortage_trend(env, lambda s, e: threshold_policy(s, e, threshold=10), "Threshold (10)"))
shortage_trends.append(evaluate_shortage_trend(env, myopic_policy, "Myopic"))

# Plotting
plt.figure(figsize=(12, 6))
for label, trend in shortage_trends:
    plt.plot(range(1, len(trend)+1), trend, label=label)

plt.xlabel("Day of Episode")
plt.ylabel("Shortage Frequency (Fraction of Episodes)")
plt.title("Shortage Trend Over Episode: DQN vs Baselines")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
env.debug = False 
def evaluate_average_inventory(env, policy_func, label, episodes=100):
    average_inventory_per_episode = []

    for _ in range(episodes):
        state = env.reset()
        done = False
        daily_inventory = []

        while not done:
            action = policy_func(state)
            state, _, done, _ = env.step(action)
            daily_inventory.append(state[0])  # inventory_level

        average_inventory_per_episode.append(np.mean(daily_inventory))

    return label, np.mean(average_inventory_per_episode), np.std(average_inventory_per_episode)

# Define policies
def fixed_policy(state):
    return env.valid_order_quantities.index(20)

def threshold_policy(state, threshold=10):
    inventory = state[0]
    return env.valid_order_quantities.index(20) if inventory < threshold else 0

def myopic_policy(state):
    inventory = state[0]
    demand_today = state[3]
    best_action = 0
    best_cost = float('inf')
    for i, order_qty in enumerate(env.valid_order_quantities):
        projected_inventory = inventory + order_qty
        shortage = max(0, demand_today - projected_inventory)
        holding = max(0, projected_inventory - demand_today)
        trucks = order_qty // env.truck_capacity
        cost = (
            shortage * env.shortage_cost +
            holding * env.storage_cost +
            trucks * env.transport_cost
        )
        if cost < best_cost:
            best_cost = cost
            best_action = i
    return best_action

def dqn_policy(state):
    with torch.no_grad():
        inventory_level = state[0]
        possible_actions = [i for i, qty in enumerate(env.valid_order_quantities) if inventory_level + qty <= env.max_inventory]
        if inventory_level >= env.max_inventory:
            possible_actions = [0]
        q_values = dqn(torch.tensor(state, dtype=torch.float32).unsqueeze(0)).squeeze(0)
        valid_q_values = [q_values[action] for action in possible_actions]
        return possible_actions[torch.argmax(torch.tensor(valid_q_values)).item()]

# Run comparison
results = []
for policy_func, label in [
    (dqn_policy, "DQN"),
    (fixed_policy, "Fixed"),
    (lambda s: threshold_policy(s, threshold=10), "Threshold (10)"),
    (myopic_policy, "Myopic")
]:
    results.append(evaluate_average_inventory(env, policy_func, label))

# Plot
labels, means, stds = zip(*results)
plt.figure(figsize=(10, 6))
bars = plt.bar(labels, means, yerr=stds, capsize=10)
bars[0].set_color('green')
plt.ylabel("Average Inventory Level")
plt.title("Average Inventory Level Comparison across Policies")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def evaluate_inventory_trend(env, policy_func, label, episodes=100):
    all_episode_inventory = []

    for _ in range(episodes):
        state = env.reset()
        done = False
        daily_inventory = []

        while not done:
            action = policy_func(state)
            state, _, done, _ = env.step(action)
            daily_inventory.append(state[0])  # inventory level

        all_episode_inventory.append(daily_inventory)

    all_episode_inventory = np.array(all_episode_inventory)  # shape (episodes, days)
    mean_inventory_by_day = all_episode_inventory.mean(axis=0)

    # Apply moving average smoothing (window size = 2)
    smoothed = np.convolve(mean_inventory_by_day, np.ones(2)/2, mode='same')

    return label, smoothed

# Reuse existing policy functions
policies = [
    (dqn_policy, "DQN"),
    (fixed_policy, "Fixed Order"),
    (lambda s: threshold_policy(s, threshold=10), "Threshold (10)"),
    (myopic_policy, "Myopic")
]

# Run evaluations
results = [evaluate_inventory_trend(env, func, name) for func, name in policies]

# Plot
plt.figure(figsize=(14, 6))
for label, smoothed_inventory in results:
    plt.plot(smoothed_inventory, label=label)

plt.title("Average Inventory Trend Over Time (Smoothed, Last 100 Episodes)")
plt.xlabel("Day")
plt.ylabel("Average Inventory Level")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
def evaluate_cumulative_cost(env, policy_func, label, episodes=100):
    all_costs = []

    for _ in range(episodes):
        state = env.reset()
        done = False
        daily_costs = []

        while not done:
            action = policy_func(state)
            state, _, done, info = env.step(action)
            daily_costs.append(info["cost"])  # daily cost from env

        cumulative_costs = np.cumsum(daily_costs)
        all_costs.append(cumulative_costs)

    all_costs = np.array(all_costs)
    mean_cumulative_costs = all_costs.mean(axis=0)
    return label, mean_cumulative_costs

# List of policies with labels
policies = [
    (dqn_policy, "DQN"),
    (fixed_policy, "Fixed Order"),
    (lambda s: threshold_policy(s, threshold=10), "Threshold (10)"),
    (myopic_policy, "Myopic")
]

# Evaluate all policies
results = [evaluate_cumulative_cost(env, func, name) for func, name in policies]

# Plot
plt.figure(figsize=(12, 6))
for label, costs in results:
    plt.plot(costs, label=label)

plt.title("Cumulative Cost Over Episode (Last 100 Episodes)")
plt.xlabel("Day")
plt.ylabel("Cumulative Cost (£)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
